```python
#Bell lepadula model for confidentiality
#Admin : Customer : Manager: Employee
#classification:- Admin:TS, Customer:Secret, Manager: Classified, Employee:Unclassified
#Categories:- Admin, Manager, Employee: G1(1); Customer: G2(0)
#No read up policy to preserve confidentiality
def createObject(H, object):
  #assign new state to new object
  #'Rewriting of Objects' property
  H.append(object)
def readData(classification, category, map, acm, subject, obj, H, f):
  if obj not in H:
    print("Non-accessibility of inactive objects voilated. Aborted.")
    return
  if f[subject][0] >= classification[obj]:
    if category[subject] >= category[obj]:
      if acm[map[subject]][map[obj]].find('r') != -1:
        print(subject, " read over ", obj, " is successfull.")
      else:
        print(subject, " read over ", obj, " is Unsuccessfull. No access.")
    else:
      print("Low category subject[",subject,"] can't read high category object[",obj,"]. Operation Aborted.")
  else:
    print("Low classification subject[",subject,"] can't read high classification object[",obj,"]. Operation Aborted.")

def editData(classification, category, map, acm, subject, obj, H, f):
  if obj not in H:
    print("Non-accessibility of inactive objects voilated. Aborted.")
    return
  if f[subject][0] >= classification[obj]:
    if category[subject] >= category[obj]:
      if acm[map[subject]][map[obj]].find('w') != -1:
        print(subject, " edit over ", obj, " is successfull.")
      else:
        print(subject, " edit over ", obj, " is Unsuccessfull. No access.")
    else:
      print("Low category subject[",subject,"] can't edit high category object[",obj,"]. Operation Aborted.")
  else:
    print("Low classification subject[",subject,"] can't edit high classification object[",obj,"]. Operation Aborted.")

#acm: access matrix
acm = [['r,w', 'r,w', 'r,w', 'r,w', 'r,w','r'],
       ['', '', '', '', '', 'r'],
       ['', 'r,w', 'r,w', 'r,w', 'r', ''],
       ['', '', 'r,w', 'r', '', '']]
#map is used to access the acm (just defining the index of acm matrix)
map = {"Admin": 0, "Customer": 1, "Manager": 2, "Employee": 3, "adminData":0, "managerData": 1, "employeeData": 2, "empSal": 3, "manSal":4, "predictionData":5}
classification = {"Admin":3, "Customer":2, "Manager":1, "Employee":0, "adminData":3, "predictionData":2, "managerData":1, "employeeData":0, "empSal":0, "manSal":1}
category = {"Admin":1, "Manager":1, "Employee":1, "adminData":1, "predictionData":0, "managerData":1, "employeeData":1, "empSal":1, "manSal":1, "Customer": 0}
#f represents clearance and current security level
f = {"Admin":[3, 1], "Customer":[2, 2], "Manager":[1, 0], "Employee":[0,0]}
#H maintains object hirearchy
H = []
createObject(H, "adminData")
createObject(H, "predictionData")
createObject(H, "managerData")
createObject(H, "employeeData")
createObject(H, "manSal")
createObject(H, "empSal")
readData(classification, category, map, acm, "Admin", "adminData", H, f)
readData(classification, category, map, acm, "Customer", "adminData", H, f)
readData(classification, category, map, acm, "Customer", "employeeData", H, f)
readData(classification, category, map, acm, "Employee", "employeeData", H, f)
readData(classification, category, map, acm, "Customer", "predictionData", H, f)
readData(classification, category, map, acm, "Manager", "predictionData", H, f)
readData(classification, category, map, acm, "Admin", "predictionData", H, f)
readData(classification, category, map, acm, "Admin", "managerData", H, f)
readData(classification, category, map, acm, "Admin", "manSal", H, f)
readData(classification, category, map, acm, "Admin", "empSal", H, f)
readData(classification, category, map, acm, "Manager", "manSal", H, f)
readData(classification, category, map, acm, "Manager", "empSal", H, f)
readData(classification, category, map, acm, "Employee", "manSal", H, f)
readData(classification, category, map, acm, "Employee", "empSal", H, f)
readData(classification, category, map, acm, "Customer", "manSal", H, f)
readData(classification, category, map, acm, "Customer", "empSal", H, f)
#ensuring integrity now
#using Biba Model to ensure integrity. Defining new classifications and categories
classification_i = {"Admin":3, "Customer":2, "Manager":1, "Employee":0, "adminData":3, "predictionData":2, "managerData":1, "employeeData":0, "empSal":1, "manSal":3}
category_i = {"Admin":1, "Manager":1, "Employee":1, "adminData":1, "predictionData":2, "managerData":1, "employeeData":1, "empSal":1, "manSal":1, "Customer": 0}
#calling various edit operations
editData(classification_i, category_i, map, acm, "Admin", "adminData", H, f)
editData(classification_i, category_i, map, acm, "Admin", "managerData", H, f)
editData(classification_i, category_i, map, acm, "Admin", "predictionData", H, f)
editData(classification_i, category_i, map, acm, "Manager", "adminData", H, f)
editData(classification_i, category_i, map, acm, "Manager", "manSal", H, f)
editData(classification_i, category_i, map, acm, "Manager", "managerData", H, f)
editData(classification_i, category_i, map, acm, "Manager", "employeeData", H, f)
editData(classification_i, category_i, map, acm, "Manager", "empSal", H, f)
editData(classification_i, category_i, map, acm, "Employee", "adminData", H, f)
editData(classification_i, category_i, map, acm, "Employee", "employeeData", H, f)
editData(classification_i, category_i, map, acm, "Employee", "empSal", H, f)
editData(classification_i, category_i, map, acm, "Customer", "predictionData", H, f)
```
