```python
def acm(acm_matrix):
    #     1  2  3  4
    #add [11, 10, '00', 10]
    #sub [10, 10, 11, 11]
    #mul [11, 10, 10, '00']
    #div [10, 11, '00', 11]
    # you can consider there are four subjects [ADD, SUB, MUL, DIV] and four OBJECTS [1, 2, 3, 4]. these 4 objects are txt files
    #acm_matrix contains the which subject has access to particular object. E.g., ADD has read-write access to object "1"
    # MUL has read access to object "2"
    num2 = int(input("Enter any file number from (1 - 4):"))
    if num2 > 4 or num2 < 1:
        print("Please Enter any number in 1-4 range.")
        return
    #now lets ask the user which operation he wants to do on these numbers
    operation = str(input("Enter any user name ('ADD' 'SUB' 'MUL' 'DIV'):"))

    # now lets perform the operation and try to see if that operation is allowed
    if operation == 'ADD':
        acm_ind_f = 0
    elif operation == 'SUB':
        acm_ind_f = 1
    elif operation == 'MUL':
        acm_ind_f = 2
    elif operation == 'DIV':
        acm_ind_f = 3
    else:
        print("Please enter the correct username.")
        return

    #now lets check acm_matrix
    #this part demonstrates the HCU(Harrison-Ruzzo-Ullman)
    print("-----------------------------")
    fileName = r'F:\M. Tech CSIS\LAB\\'+str(num2)+".txt"
    if acm_matrix[acm_ind_f][num2-1] == '00':
        print("The ",operation," is not allowed to read from or write to the file ",num2,".txt")
    else:
        if acm_matrix[acm_ind_f][num2-1] == 10:
            print("The ",operation," is allowed to read from but not write to the file ",num2,".txt")
        elif acm_matrix[acm_ind_f][num2-1] == 11:
            print("The ",operation," is allowed to read from & write to the file ",num2,".txt")
            choice = input("Do you want to write something(Y/N):")
            if choice == 'Y' or choice == 'y':
                data = input("Enter data:")
                f = open(fileName, "a")
                f.write(data)
                f.close()
        f = open(fileName, "r")
        print("------File Data--------")
        print(f.read())
        f.close()


#function to print capability lists
def cap_list(acm_matrix):
    print("-----------------------------\nCapability Lists:")
    m = len(acm_matrix)
    n = len(acm_matrix[0])
    for i in range(0, m):
        cp_list = []
        if i == 0:
            op = "ADD"
        elif i == 1:
            op = "SUB"
        elif i == 2:
            op = "MUL"
        elif i == 3:
            op = "DIV"
        for j in range(0, n):
            if acm_matrix[i][j] != '00':
                cp_list.append([op,j+1, acm_matrix[i][j]])
        print(cp_list)
        
#function to print access control lists
def ac_list(acm_matrix):
    print("-----------------------------\nAccess Control Lists:")
    m = len(acm_matrix)
    n = len(acm_matrix[0])
    for i in range(0, n):
        cp_list = []
        for j in range(0, m):
            if j == 0:
                op = "ADD"
            elif j == 1:
                op = "SUB"
            elif j == 2:
                op = "MUL"
            elif j == 3:
                op = "DIV"
            if acm_matrix[j][i] != '00':
                cp_list.append([i+1, op, acm_matrix[j][i]])
        print(cp_list)
        
#main driver code
matrix = [[11, 10, '00', 10], [10, 10, 11, 11], [11, 10, 10, '00'], [10, 11, '00', 11]]
#00 means no access, 01 means only write access, 10 means only read, 11 means read-write access
acm(matrix)
cap_list(matrix)
ac_list(matrix)

```
