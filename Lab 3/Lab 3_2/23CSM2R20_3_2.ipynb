```python
#function to print capability lists
def cap_list(acm_matrix, users, objects):
    m = len(acm_matrix)
    cp_list_final = []
    for i in range(0, m):
        n = len(acm_matrix[i])
        cp_list = []
        for j in range(0, n):
            if acm_matrix[i][j] != '':
                cp_list.append([users[i],objects[j], acm_matrix[i][j]])
        cp_list_final.append(cp_list)
    return cp_list_final
        
#function to print access control lists
def ac_list(acm_matrix, users, objects):
    m = len(acm_matrix)
    n = len(acm_matrix[0])
    cp_list_final = []
    for i in range(0, n):
        cp_list = []
        for j in range(0, m):
            if i<len(acm_matrix[j]) and acm_matrix[j][i] != '':
                cp_list.append([objects[i], users[j], acm_matrix[j][i]])
        cp_list_final.append(cp_list)
    return cp_list_final

#function to create a file
def createObject(name, objects, access_control_list):
    filename = input("Enter filename:")
    data = input("Enter data:")
    fileName = filename+".txt"
    f = open(fileName, "w")
    f.write(data)
    f.close()
    #add filename to objects list
    objects[len(objects)] = filename
    access_control_list.append([filename, users[name], "rwxo"])
#function to delete a file
def deleteObject(username, objects, access_control_list):
    import os
    filename = input("Enter filename:")
    if filename not in objects.values():
        print("File does not exist")
        return
    fileName = filename+".txt"
    #check access rights then delete
    name = list(objects.keys())[list(objects.values()).index(filename)]
    n = len(access_control_list[name])
    for i in range(0,n):
        if access_control_list[name][i][1] == users[username] and access_control_list[name][i][2]=="rwxo":
            os.remove(fileName)
            print("Deleted the object")
            access_control_list.pop(name)
            objects.pop(name)
            return
    print("You dont have delete access.")
    
#function to read from a file
def read(username, objects, access_control_list):
    filename = input("Enter filename:")
    if filename not in objects.values():
        print("File does not exist")
        return
    fileName = filename+".txt"
    #check read access
    name = list(objects.keys())[list(objects.values()).index(filename)]
    n = len(access_control_list[name])
    for i in range(0,n):
        print(access_control_list[name][i][1],access_control_list[name][i][2])
        if access_control_list[name][i][1] == users[username] and access_control_list[name][i][2].find('r') != -1:
            f = open(fileName, "r")
            print("------File Data--------")
            print(f.read())
            f.close()
            return
    print("You don't have read access.")
    
#function to write to a file
def write(username, objects, access_control_list):
    filename = input("Enter filename:")
    if filename not in objects.values():
        print("File does not exist")
        return
    fileName = filename+".txt"
    #check read access
    name = list(objects.keys())[list(objects.values()).index(filename)]
    n = len(access_control_list[name])
    for i in range(0,n):
        print(access_control_list[name][i][1],access_control_list[name][i][2])
        if access_control_list[name][i][1] == users[username] and access_control_list[name][i][2].find('w') != -1:
            data = input("Enter data:")
            f = open(fileName, "a")
            f.write(data)
            f.close()
            print(fileName+" Saved successfully.")
            return
    print("You don't have write access.")

    
#main driver code
#matrix is an ACM matrix
#we will define 4 parameters: read | write | execute | ownership ; represented as r,w,x,o respectively
matrix = [['rwxo', 'r','', 'r'], ['r', 'rw', 'rw', 'rwxo'], ['rw', 'r', 'rwxo', ''], ['r', 'rwxo', '', 'rw']]
capabality_list = cap_list(matrix, users, objects)
access_control_list = ac_list(matrix, users, objects)
#maintain a list of users using dictionary
users = {0:"user1", 1:"user2", 2:"user3", 3:"user4"}
#objects
objects = {0:"file1", 1:"file2", 2:"file3", 3:"file4"}

username = input("Enter your username:")
name = list(users.keys())[list(users.values()).index(username)]
#printing the acm , cl , and acl
print("-----------------------------\nAccess Control Matrix1:\n",matrix)
print("-----------------------------\nAccess Control Lists:\n",access_control_list)
print("-----------------------------\nCapability Lists:\n",capabality_list)
while 1:
    if username not in users.values():
        print("This username doesn't exist.")
        break
    print("Enter\n1 to generate access list\n2 to create an object\n3 to delete an object\n4 read from a file\n5 write to a file\n6 to exit.")
    choice = int(input("Your choice: "))
    if choice == 1:
        print(access_control_list)
    elif choice == 2:
        createObject(name, objects, access_control_list)
    elif choice == 3:
        deleteObject(name, objects, access_control_list)
    elif choice == 4:
        read(name, objects, access_control_list)
    elif choice == 5:
        write(name, objects, access_control_list)
    if choice == 6:
        print("Bye...")
        break  

```
